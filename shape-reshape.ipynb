## There are 2 very relevant functions on arrays (Shape and Reshape). 

The shape of an array will tell us what structure we are working with (size, manipulate, enter).

# Shape

Indicates the shape of the arrangement.

```python
import numpy as np
```

```python
arr = np.random.randint(1,10, (3,2))
arr.shape
arr
```

# Reshape

Transforms the array as long as the elements reamin as maintained.

```python
arr.reshape(1,6)
```

```python
arr.reshape(2,3)
```

```python
np.reshape(arr,(1,6))
```

### You can perform a reshape as C would.

```python
np.reshape(arr,(2,3), 'C')
```

### You can perform a reshape just like Fortran would.

```python
np.reshape(arr,(2,3), 'F')
```
