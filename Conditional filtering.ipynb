# Conditional Filtering

---
### It works by boolean logic and returns the values that are in "True". It is very useful because sometimes we want to filter or separate data.


```python
import numpy as np
import pandas as pd
```

### Let's load data 
We call the data from a csv file to handle it

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv', sep=',',header=0)
df_books
```

### Display data that is greater than a certain value

```python
greater_2016 = df_books['Year'] > 2016
greater_2016
```

### Filter data in our DataFrame that is greater than a certain value

```python
df_books[greater_2016]
```

### You could also place the condition directly as a parameter

```python
df_books[df_books['Year'] > 2016]
```

### Display data that is equal to a certain value

```python
GenreFiction = df_books['Genre'] == 'Fiction'
GenreFiction
```

### Filtering with multiple conditions

```python
df_books[GenreFiction & greater_2016]
```

### Disavowal filtering

```python
df_books[~greater_2016]
```
