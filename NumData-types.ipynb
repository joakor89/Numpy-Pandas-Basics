NumPy arrays can only contain one data type, as this is what gives it the benefits of memory optimization.

We can know the data type of the array by consulting the .dtype property

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4])
arr.dtype
```

By adding the dtype float to the array, lets makes it easier to work with neural networks

```python
arr = np.array([1,2,3,4], dtype = 'float64')
arr.dtype
```

```python
arr
```

```python
arr = np.array([1,2,3,4])
arr.dtype
```

```python
arr = arr.astype(np.float64)
```

```python
arr = np.array([1,2,3,4])

arr = arr.astype(np.bool_)
arr
```

On booleans logic, zero will always be represented as False and the rest of number, Python & NumPy will represent them as True.

```python
arr = np.array([1,2,3,4])

arr = arr.astype(np.string_)
arr
```

the np.string_ let NumPy known by know, the integer will be manage as strings

```python
arr = np.array(['1','2','3','4'])

arr = arr.astype(np.int8)
arr
```
