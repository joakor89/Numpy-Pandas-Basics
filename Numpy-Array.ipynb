### Slicing



1.   It allows us to extract various data elements and has a beginning and an end.
2.   Numpy will read empty spaces as "all or whole" from the start or the end by convention when slicing is performed. In other words, arra[:2] it's going to be read from the start until index element 2.



---
arr[:] it will read everything.

arr[::3] It'll read everything but read it three by three.

```python
arr
```

```python
arr[1:6]
```

```python
arr[:2]
```

```python
arr[::3]
```

Any value represented by a minus sign(-) will start the indexing from the end to the beginning of the array.

```python
arr[-1]
```

```python
arr[-3:]
```

```python
matrix
```

```python
matrix[1:,0:2]
```
