# Copy 

---
| .copy() | allows us to copy a NumPy array into another variable in such a way that when modifying the new array the changes are not reflected in the original array.


```python
import numpy as np
```

```python
arr = np.arange(0,11)
arr
```

Let's take a piece from the original array

```python
arr_piece = arr[0:6]
arr_piece
```

```python
arr_piece[:] = 0
arr_piece
```

```python
arr
```

```python
arr_copy = arr.copy()
```

```python
arr_copy[:] = 100
```

```python
arr_copy
```

```python
arr
```

The data in the original array has been modified because it was still referencing that variable.

With .copy() we create a copy so as not to damage our original array
