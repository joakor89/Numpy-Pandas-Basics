

*     A Python list understands that you want to duplicate the data. It's not what we're looking for. However Numpy understands it much better

Let's perform it on NumPy:



```python
arr = np.arange(0,10)
```

```python
arr2 = arr.copy()
```

```python
arr * 2
```

```python
arr + 2
```

```python
1 / arr
```

```python
arr**2
```

```python
arr + arr2
```

## Let's check it out on Matrix

```python
matrix = arr.reshape(2,5)
```

```python
matrix2 = matrix.copy()
```

```python
matrix
```

```python
matrix + matrix2
```

```python
matrix - matrix2
```

```python
matrix2.T
```

```python
np.matmul(matrix, matrix2.T)
```

```python
matrix @ matrix2.T
```
