# Conditions 

---
Allow us to look up into more specific consults


```python
import numpy as np
```

```python
arr = np.linspace(1,10,10, dtype='int8')
arr
```

### Conditionals will returns boolean values for where the conditions is True.

```python
cond_index = arr > 5
cond_index
```

```python
arr[cond_index]
```

### Multiple conditions

```python
arr[(arr > 5)&(arr < 9)]
```

```python
arr[arr > 5] = 99
arr
```
