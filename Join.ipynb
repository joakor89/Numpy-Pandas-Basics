# Join 
---
*   It's a tool which perform the same thing as merge, a combination. 
*   The difference with merging is that join is going to focus directly on indexes and rather than any specific columns.


```python
import pandas as pd
import numpy as np
```

### Join

Join will look for an index match.

```python
left = pd.DataFrame({'A': ['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                    index=['k0','k1','k2'])

right =pd.DataFrame({'C': ['C0','C1','C2'],
                     'D':['D0','D1','D2']},
                    index=['k0','k2','k3']) 
```

### Let's gather both DataFrames

```python
left.join(right)
```

Let's now apply "how=" to guide the way joins are going to work on

```python
left.join(right, how='inner')
```

```python
left.join(right, how='right')
```

```python
left.join(right, how='left')
```

```python
left.join(right, how='outer')
```
