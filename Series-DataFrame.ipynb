# Pandas series

---
## It is very similar to a NumPy one-dimensional array (or vector).

*   Indexed one-dimensional array
*   Search by index
*   Slicing
*   Arithmetic operations
*   Different types of data

# Pandas DataFrame

---
## Very similar to the matrix structures worked with NumPy.

*   Main structure
*   Two-dimensions array
*   Search by index (columns or rows)
*   Slicing
*   Arithmetic operations
*   Different types of data
*   Variable size

```python
import pandas as pd
```

### Defining a pandas series with a list

```python
psg_players=pd.Series(['Navas', 'Mbappe', 'Neymar', 'Messi'], 
          index = [1,7,10,30]
          )
```

```python
psg_players
```

### Defining Pandas series with a dictionary

```python
psg_dict = {1:'Navas', 7:'Mbappe', 10:'Neymar', 30:'Messi'}
pd.Series(psg_dict)
```

```python
psg_players[7]
```

```python
new_dict = {'Player':['Navas', 'Mbappe', 'Neymar', 'Messi'],
 'Height': [1.85, 178.4, 172.7, 170.0],
 'Scored_Goals':[ 2, 259, 434, 790],
 }
```

```python
df_players = pd.DataFrame(new_dict, index = [1,7,10,30])
df_players
```

```python
df_players.columns
```

```python
df_players.index
```
