# Apply 

---
*   It's a very powerful command that allows us to apply functions to our DataFrame

Let's import and set up the DataFrame

```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv', sep=',', header=0)
df_books
```

```python
df_books.head(5)
```

Let's create a function

```python
def two_times(value):
  return value * 2
```

Let's now applied to the User Rating column

```python
df_books['User Rating'].apply(two_times)
```

Let's now gather & save al this new information within a new column

```python
df_books['User Rating2'] = df_books['User Rating'].apply(two_times)
df_books
```

You can perform the same procedure with a Lambda function

```python
df_books['User Rating2'] = df_books['User Rating'].apply(lambda x: x*3)
df_books.head(5)
```

```python
df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis = 1)
df_books.head(3)
```
