### Null Data Handling

---
Null data is a headache on Data Science's world and it's going to be found several times in our DataFrames

### We create a DataFrame with some null values

```python
import pandas as pd
import numpy as np
```

```python
dict = {'Col1':[1,2,3,np.nan],
'Col2':[4, np.nan,6,7],
'Col3':['a','b','c', None]}
```

```python
df = pd.DataFrame(dict)
df
```

### Identifying null values in a DataFrame

```python
df.isnull
```

### Identify null values with a numeric value

```python
df.isnull() * 1
```

### Replace null values with a string

```python
df.isnull('Missing')
```

### Replace null values with statistical measurement made with the values of the columns

```python
df.fillna(df.mean())
```

### Replace null values with interpolated values

```python
df.interpolate()
```

### Remove null values

```python
df.dropna()
```
