# Iloc & loc 

---

*   Used to navigate a DataFrame.
*   These functions allow you to filter data more specifically


```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv', sep=',', header=0, names=['Name', 'Author', 'User Rating', 'Reviews', 'Price', 'Year', 'Genre'])
```

```python
df_books.columns
```

```python
df_books
```

```python
df_books.loc[0:4]
```

```python
df_books.loc[0:4, ['Name','Author']]
```

```python
df_books.loc[:, ['Reviews']] * -1
```

```python
df_books.loc[:, ['Author']] == 'JJ Smith'
```

```python
df_books.iloc[:]
```

```python
df_books.iloc[:4, 0:2]
```

```python
df_books.iloc[1,3]
```

```python
df_books.iloc[:5, 2:]
```
